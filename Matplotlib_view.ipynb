### Знакомство с matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.cm import get_cmap
import random
```

```python
data_x1 = [k for k in range(10)]
data_y1 = [j-5 for j in range(10)]
```

```python
plt.plot(data_x1, data_y1)
plt.show()
```

```python
random.sample(range(100), 10)
```

```python
data_x2 = random.sample(range(30), 20)
data_y2 = random.sample(range(-50, 50, 1), 20)
data_x2.sort()
data_y2.sort()
```

```python
plt.figure()
plt.plot(data_x2, data_y2, color='red')
plt.figure()
plt.plot(data_x1, data_y1)
plt.show()
```

```python
with open('data_set//data-398-2019-11-21.csv', 'r', encoding='cp1251') as fd:
    stations = pd.read_csv(fd, sep=';')
```

```python
stations = stations[stations['District'] != 'District']
```

```python
stations.groupby(['District']).size()
```

```python
admarea_count = stations.groupby(['AdmArea']).size()
admarea_count
```

```python
labels = ['ВАО','ЗАО', 'Зел', 'Нов', 'САО', 'СВАО', 'СЗАО', 'Тро', 'ЦАО', 'ЮВАО', 'ЮЗАО', 'ЮАО' ]
```

```python
# выбор цветовой схемы
cmap = get_cmap('tab20')
colors = [cmap(i) for i in range(len(labels))]
```

```python
plt.title('Остановки по округам Москвы')
plt.bar(labels, admarea_count, color=colors)
plt.show()
```
