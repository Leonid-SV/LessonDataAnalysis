```python
embarked = df.groupby(['Embarked']).count()['PassengerId']
```

```python
embarked_max = embarked[embarked == embarked.max()].index
```

```python
df.loc[df['Embarked'].isnull(), ['Embarked']] = embarked_max
```

```python
df[df.Embarked.isnull()]
```

Получение титула из колонки Name

```python
df['Title'] = df['Name'].str.extract(r'([A-Z]+[a-z]+)\.')
```

```python
df.columns
```

```python
df['Title'].unique()
```

```python
mapping = {'Don': 'RoyaltyM', 'Mme': 'Miss', 'Ms': 'Miss', 'Major': 'Officer', 
           'Lady': 'RoyaltyF', 'Sir': 'RoyaltyM', 'Mlle': 'Miss', 'Col': 'Officer', 
           'Capt': 'Officer', 'Countness': 'RoyaltyF', 'Jonkheer': 'RoyaltyM'}
```

```python
df.replace({'Title': mapping}, inplace=True)
```

```python
df
```

```python
age_med = df.groupby('Title')['Age'].median()
```

```python
age_med
```

```python
for title, age in age_med.iteritems():
    filter_age = df['Age'].isnull()
    filter_title = df['Title'] == title
    df.loc[filter_age & filter_title, 'Age'] = age
```

```python
df.info()
```

Замена Male / Female на 1 / 0 соответственно

```python
mapping_sex = {'male': 1, 'female': 0}
```

```python
df.replace({'Sex': mapping_sex}, inplace=True)
```

```python
survived_pclass = df.pivot_table(values='Name', index='Pclass', columns='Survived', aggfunc='count')
survived_title = df.pivot_table(values='Name', index='Title', columns='Survived', aggfunc='count')
```

```python
survived_pclass.plot(kind='bar', stacked=True)
```

```python
survived_title.plot(kind='bar', figsize=(15, 10))
```
